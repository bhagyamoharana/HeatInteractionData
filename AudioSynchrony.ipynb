```python
import os
from pydub import AudioSegment
from pyannote.audio import Pipeline
import matplotlib.pyplot as plt

# === CONFIGURATION ===
audio_path = r"C:\Users\bmoha\Desktop\DataCollection\Dublin\1455\Audio\audiofile.mp3"
output_dir = os.path.dirname(audio_path)
diarization_plot = os.path.join(output_dir, "speaker_diarization.png")

# === Load audio ===
print("ðŸ”Š Loading audio...")
audio = AudioSegment.from_mp3(audio_path)

# === Diarization pipeline (pre-trained model) ===
print("ðŸ§  Running speaker diarization (may take time)...")
pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization")

# Apply diarization (can take time)
diarization = pipeline(audio_path)

# === Print segments ===
print("\nðŸ“‹ Diarization Results (who spoke when):")
for turn, _, speaker in diarization.itertracks(yield_label=True):
    print(f"{speaker}: {turn.start:.1f}s - {turn.end:.1f}s")

# === Plotting ===
print("\nðŸ“Š Plotting speaker timeline...")
fig, ax = plt.subplots(figsize=(12, 3))
for turn, _, speaker in diarization.itertracks(yield_label=True):
    ax.plot([turn.start, turn.end], [speaker, speaker], lw=6)
ax.set_title("Speaker Diarization Timeline")
ax.set_xlabel("Time (s)")
plt.tight_layout()
plt.savefig(diarization_plot, dpi=300)
plt.show()

print(f"\nâœ… Saved diarization plot to: {diarization_plot}")
```
